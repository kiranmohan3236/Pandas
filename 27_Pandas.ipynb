```python
from numpy.random import randn
```

```python
np.random.seed(101) # seed - to get the same random numbers
```

```python
df['new'] = df['W'] + df['Y']
```

```python
df
```

```python
#deleting the 'new' column
df.drop('new',axis=1)
```

```python
#without the inplace = True
df #df has the new not being deleted
```

```python
df.drop('new',axis=1,inplace=True)
#inplace = True if we want it to be permanent
```

```python
df #now using the inplace = True, new is deleted
```

```python
#drop 'E' row
df.drop('E')  #df.drop('E',axis=0)
```
