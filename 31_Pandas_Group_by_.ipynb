```python
#groupby allows you to group together rows based off of a column and perform an aggregate function on them.
```

```python
import numpy as np
import pandas as pd
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
        'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
        'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
byComp = df.groupby('Company')
```

```python
byComp
```

```python
byComp.mean()
```

```python
byComp.sum()
```

```python
byComp.std()
```

```python
byComp.sum().loc['FB']
```

```python
df.groupby('Company').sum().loc['FB']
```

```python
df.groupby('Company').count()
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').min()
```

```python
df.groupby('Company').describe()
```

```python
df.groupby('Company').describe().transpose()
```

```python
df.groupby('Company').describe().transpose()['FB'] #same output
```

```python
df.groupby('Company').describe().loc['FB'] #same output
```
