```python
import numpy as np
import pandas as pd
```

```python
d = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]} #np.nan for null values
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df.dropna(axis = 1) #drop - delete
```

```python
df.dropna() #drop - delete
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=2)
```

```python
#fillna - fill missing values
```
